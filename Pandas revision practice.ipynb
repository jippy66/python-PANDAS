```python
def ages(age):
    if age < 10:
        return "category_1"
    elif age < 30:
        return "category_2"
    elif age < 60:
        return "category_3"
    elif age <=80:
        return "category_4"
```

```python
df["age"]=df["age"].apply(ages)
```

```python
# ans 2
dk=df.groupby("age").agg("mean")
dk
```
